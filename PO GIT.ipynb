#####  'PO Type' identify whether this PO is a sudden strike or a planned power-off for maintenance. The zip code of all data starts with '90 ', which means that it is located in the administrative district of Ghent.  
##### The  ‘GHENT’ in the 'Location' column indicates that this one is in the Ghent city area (corresponding to the zip code 9000).  
#####  Date (dd-mm-yyyy) indicate the day when this PO occurs. 'StartH' and 'StartM' indicate the start time of the PO. e.g. 'StartH' : 8 + 'StartM': 30 means this PO start at 08:30 that day. Same rule applies to 'EndH' and 'EndM'


```python
# Transfer the time into the pd format and sort it with start time
df = df_RawData
df['datetime'] = pd.to_datetime(df['Date'] + ' ' + df['StartH'].astype(str) + ':' + df['StartM'].astype(str), format='%d-%m-%Y %H:%M')
df_sorted = df.sort_values(by='datetime')
df_PO = df_sorted

#present the processed data, with time sequence
df_PO 
```

###### now the 'datetime' column indicate the absolute start time of the PO

```python
# Calculate the duration time
df_PO = df

# Use the absolute end minute to minus the absolute start minute， get the 'duration' in absolute minutes
df['duration']=(df['EndH']*60 + df['EndM'])-(df['StartH']*60 + df['StartM'])

# if the 'duration' is negative, meaning that the a PO start on night of a day, end on next day, pass the midnight.
# the negative value + 24*60 should be the absolute minutes (Assume no PO can pass two consective midnights)
df.loc[df['duration'] < 0, 'duration'] += 60 * 24 

# transfer absolute min to abosulyte hour, for better view
df['duration'] = df['duration']/60
df = df_PO

# present the data 
df_PO 
```

#### The collected data only indicate the impacted user in a range '<20', '21-100','101-500','501-1500'. 
#### We need to specify a impacted user number, several options can be used, for example:
####  (1) random a integer within the range with no other conditions
####  (2) random a integer within the upper side of the range (e.f. for '21-100', random the integer from 61-100) 
####      if the occupants are vulnerable(elder,children, patient), we may want to slightly overestimate the PO impact to make sure the health and wellbeing
####  (3) random a integer within the range, but try to make sure all the integers for each range are in normal distribution.
####  (4)... more rules can be defined by the yourself according to your own needs.
#### in this example, we take the upper limit of the 'impact' range as the mean number of the 'impact user' creating the normal distribution numbers. For the data provided by Fluvius is not the exact user number but an estimation based on the location and address.

```python
# Add a blank column called 'Impact users'
df_PO_Sud['Impact users'] = np.nan
```

#### As for ['Impact'] showing '<20'，claculate total rows(n). Construct a normal distribution dataset consisting n numbers, each number between[1,20], randomly assign these numbers to['Impact users']. This is because the raw data from fluvius only gives impact range instead of specific number, if your own raw data has a the exact impact users, this step is not necessary.

```python
def ImpactUserNumberNorDis(df = df_PO_Sud,IPT = '<20',Range_L = 1,Range_U = 40,STD = 10,MEAN = 20,boxbar =1):
    fil_df = df[df['Impact'] == IPT]
    n = len(fil_df)
    normal_dist_data = np.random.normal(loc=MEAN, scale=STD, size=n).astype(int)
    normal_dist_data = np.clip(normal_dist_data, Range_L, Range_U)
    # Assign these integers to the 'Impact users' column of the filtered rows
    df.loc[df['Impact'] == IPT , 'Impact users'] = normal_dist_data
    
    # Plot the distribution of these integers
    plt.hist(normal_dist_data, bins=range(Range_L, Range_U), edgecolor='black')
    plt.xlabel('Impact users')
    plt.ylabel('Frequency')
    plt.title('Distribution of Impact users')
    plt.xticks(range(Range_L, Range_U,boxbar))
    plt.show()
    
    return df
```

```python
df_PO_Sud = ImpactUserNumberNorDis(df = df_PO_Sud,IPT = '<20',Range_L = 1,Range_U = 40,STD = 5,MEAN = 20,boxbar =5)
```

### do the same for other ['Impact']

```python

df_PO_Sud= ImpactUserNumberNorDis(df = df_PO_Sud,IPT = '21-100' ,Range_L=21,Range_U=180,STD = 10 ,MEAN =100,boxbar =10)
df_PO_Sud = ImpactUserNumberNorDis(df = df_PO_Sud,IPT = '101-500' ,Range_L=101,Range_U=900,STD = 100 ,MEAN =500,boxbar =100)
df_PO_Sud = ImpactUserNumberNorDis(df = df_PO_Sud,IPT = '501-1500' ,Range_L=501,Range_U=2500,STD = 300 ,MEAN =1500,boxbar =500)
```

# RMC simulation

```python
df = df_PO_Sud

# Take a random value (dice) [0,1]
df['dice'] = np.random.rand(len(df))

# Test whether the dice is equal to or lower than Prob, true:1(PO impact target building), false:0 (PO dont impact target building)
df['PO'] = np.where(df['dice'] <= df['prob'], 1, 0)

# taking out the POs that impacted the target building
df_prof = df[df['PO'] == 1]
df_prof
```

```python
# a function to generate the Profile
def PO2Profile(df = df_PO_Sud):
    
    # Take a random value (dice) [0,1]
    df['dice'] = np.random.rand(len(df))
    
    # Test whether the dice is equal to or lower than Prob, true:1(PO impact target building), false:0 (PO dont impact target building)
    df['PO'] = np.where(df['dice'] <= df['prob'], 1, 0)
    
    # filter out the POs that impacted the target building
    df_prof = df[df['PO'] == 1]
    
    # combine the filtered POs into the profile
    return df_prof
```

#### test the PSL by simulation, with different number of profiles

```python
# this part of code will consum quite a long time, depending your input PO data, simulation runs, etc

# start with 10 profiles, stop with 3000, adding 2 more each time (10,12,14........3000)
x_values = np.arange(10, 3000, 2)

# save the results
results = []

# calculate the  P_PO(by simulation) with different profiles(x)
for x in x_values:
    profile_dict = {}
    
    # generate x Profile_9000
    for i in range(x):
        profile_key = f'Profile_9000_{i+1}'  
        # run the function POtoProfile to generate profiles
        profile_dict[profile_key] = PO2Profile(df = df_PO_Sud)

    # save all profiles into dfs
    profile_dfs = [profile_dict[key] for key in profile_dict]
    # contact them into one
    combined_profile_df = pd.concat(profile_dfs, ignore_index=True)


    # drop the empty DataFrame (profile with no PO impacted target building)
    non_empty_dfs = [df for df in profile_dfs if not df.empty]

    # calculate P_PO, Profiles with PO impact target building / total profiles (x)
    P_PO = len(non_empty_dfs) / x
    # P_PO = len(combined_profile_df)/len(non_empty_dfs)
    PSL = 1-P_PO
    
    #calculate the mean duration of the simulated PO 
    Mean_Dur = combined_profile_df['duration'].mean() 
    
    print(f"x = {x},finished")

    # save the major results
    results.append((x,P_PO,PSL,Mean_Dur))

# results to DataFrame
results_df = pd.DataFrame(results, columns=['x','P_PO','PSLs','Mean_Dur'])
```
